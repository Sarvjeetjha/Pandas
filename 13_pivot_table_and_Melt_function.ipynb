                        #Melt()

```python
import pandas as pd
# changing index no of data
var1=pd.DataFrame({'days':[1,2,3,4,5,6,7],'eng':[11,12,13,14,16,12,11],'maths':[17,18,19,14,13,12,16]})
# var2=pd.DataFrame({'C':[10,30],'B':[11,22]},index=['a','b'])
var1
```

```python
pd.melt(var1)
```

```python
pd.melt(var1,id_vars=['eng'])
```

```python
pd.melt(var1,id_vars=['days'],var_name='python')
```

```python
pd.melt(var1,id_vars=['days'],var_name='python',value_name='Sarvjeet')
```

                                        Pivot()

```python
import pandas as pd
# changing index no of data
var1=pd.DataFrame({'days':[1,2,3,4,5,6,7],'std_name':['a','b','c','d','e','f','g'],'eng':[11,12,13,14,16,12,11],'maths':[17,18,19,14,13,12,16]})
# var2=pd.DataFrame({'C':[10,30],'B':[11,22]},index=['a','b'])
var1
```

```python
var1.pivot(index='days',columns='std_name')
```

```python
var1.pivot(index='days',columns='std_name',values='eng')
```

```python
var1.pivot(index='days',columns='std_name')
```

```python
import pandas as pd
# changing index no of data
var1=pd.DataFrame({'days':[1,1,1,2,2,2,2],'std_name':['a','b','c','d','e','f','g'],'eng':[11,12,13,14,16,12,11],'maths':[17,18,19,14,13,12,16]})
# var2=pd.DataFrame({'C':[10,30],'B':[11,22]},index=['a','b'])
var1
```

```python
var1.pivot_table(index='std_name',columns='days',aggfunc="sum")
```

```python
var1.pivot_table(index='std_name',columns='days',aggfunc="mean")
```

```python
var1.pivot_table(index='std_name',columns='days',aggfunc="mean",margins=True)
```
