                        Merge

```python
import pandas as pd
csv1=pd.DataFrame({"A":[1,2,3,4],'B':[11,12,13,14]})
csv2=pd.DataFrame({"A":[1,2,3,4],'C':[21,22,23,24]})
pd.merge(csv1,csv2,on='A')  # on shows commmon column of both dataframe
```

```python
import pandas as pd
csv1=pd.DataFrame({"A":[1,2,3,4],'B':[11,12,13,14]})
csv2=pd.DataFrame({"A":[1,2,3,5],'C':[21,22,23,24]})
pd.merge(csv2,csv1,on='A')  # on shows commmon column of both dataframe
# it will show only common data in both dataframe
```

```python
import pandas as pd
csv1=pd.DataFrame({"A":[1,2,3,4],'B':[11,12,13,14]})
csv2=pd.DataFrame({"A":[1,2,3,5],'C':[21,22,23,24]})
pd.merge(csv2,csv1,how='inner')  # on shows commmon column of both dataframe
# it will show only common data in both dataframe
```

```python
import pandas as pd
csv1=pd.DataFrame({"A":[1,2,3,4],'B':[11,12,13,14]})
csv2=pd.DataFrame({"A":[1,2,3,5],'C':[21,22,23,24]})
pd.merge(csv2,csv1,how='left')  # on shows commmon column of both dataframe
# it will show only common data in both dataframe
```

```python
import pandas as pd
csv1=pd.DataFrame({"A":[1,2,3,4],'B':[11,12,13,14]})
csv2=pd.DataFrame({"A":[1,2,3,5],'C':[21,22,23,24]})
pd.merge(csv2,csv1,how='right')  # on shows commmon column of both dataframe
# it will show only common data in both dataframe
```

```python
import pandas as pd
csv1=pd.DataFrame({"A":[1,2,3,4],'B':[11,12,13,14]})
csv2=pd.DataFrame({"A":[1,2,3,5],'C':[21,22,23,24]})
pd.merge(csv2,csv1,how='outer')  # on shows commmon column of both dataframe
# it will show only common data in both dataframe
```

```python
import pandas as pd
csv1=pd.DataFrame({"A":[1,2,3,4],'B':[11,12,13,14]})
csv2=pd.DataFrame({"A":[1,2,3,5],'C':[21,22,23,24]})
pd.merge(csv2,csv1,how='outer',indicator=True)  # on shows commmon column of both dataframe
# it will show only common data in both dataframe
```

```python
import pandas as pd
csv1=pd.DataFrame({"A":[1,2,3,4],'B':[11,12,13,14]})
csv2=pd.DataFrame({"A":[1,2,3,5],'B':[21,22,23,24]})
pd.merge(csv1,csv2,left_index=True,right_index=True)  # on shows commmon column of both dataframe
# it will show only common data in both dataframe
```

```python
import pandas as pd
csv1=pd.DataFrame({"A":[1,2,3,4],'B':[11,12,13,14]})
csv2=pd.DataFrame({"A":[1,2,3,5],'B':[21,22,23,24]})
pd.merge(csv1,csv2,left_index=True,right_index=True,suffixes=('name','Python'))  # on shows commmon column of both dataframe
# it will show only common data in both dataframe
```

                        Concate

```python
import pandas as pd
csv1=pd.Series([1,2,3,4])
csv2=pd.Series([21,22,23,24])

pd.concat([csv1,csv2])
```

```python
import pandas as pd
D1=pd.DataFrame({"A":[1,2,3,4],'B':[11,12,13,14]})
D2=pd.DataFrame({"A":[1,2,3,5],'C':[21,22,23,24]})
pd.concat([D1,D2])
```

```python
import pandas as pd
D1=pd.DataFrame({"A":[1,2,3,4],'B':[11,12,13,14]})
D2=pd.DataFrame({"A":[1,2,3,5],'C':[21,22,23,24]})
pd.concat([D1,D2],axis=0)
```

```python
import pandas as pd
D1=pd.DataFrame({"A":[1,2,3,4],'B':[11,12,13,14]})
D2=pd.DataFrame({"A":[1,2,3,5],'C':[21,22,23,24]})
pd.concat([D1,D2],axis=1)
```

```python
import pandas as pd
D1=pd.DataFrame({"A":[1,2,3,4],'B':[11,12,13,14]})
D2=pd.DataFrame({"A":[1,3,5],'C':[21,23,24]})
pd.concat([D1,D2],axis=1,join='inner')# inner mean intersection i:e common 
```

```python
import pandas as pd
D1=pd.DataFrame({"A":[1,2,3,4],'B':[11,12,13,14]})
D2=pd.DataFrame({"A":[1,3,5],'C':[21,223,24]})
pd.concat([D1,D2],axis=1,join='outer')# outer mean union i:e all 
```

```python
import pandas as pd
D1=pd.DataFrame({"A":[1,2,3,4],'B':[11,12,13,14]})
D2=pd.DataFrame({"A":[1,3,5],'C':[21,223,24]})
pd.concat([D1,D2],axis=0,keys=["D1","D2"])# inner mean intersection i:e common 
```

```python
import pandas as pd
D1=pd.DataFrame({"A":[1,2,3,4],'B':[11,12,13,14]})
D2=pd.DataFrame({"A":[1,3,5],'C':[21,223,24]})
pd.concat([D1,D2],axis=1,keys=["D1","D2"])# inner mean intersection i:e common 
```
