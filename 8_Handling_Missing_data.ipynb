                        Handling Missing Data

                            dropna & fillna

                            Note;-axis =0->>row
                                axis=1,->>column

```python

import pandas as pd
csv=pd.read_csv('IRIS.csv')    #reading csv file
csv.loc[[2,3,4,20],['petal_length','species']]=None
csv
```

```python
csv.dropna()# to drop NaN containing row
```

```python
csv.dropna(axis=0)# to drop NaN along row
```

```python
csv.dropna(axis=1)# to drop NaN along coloumn
```

```python
csv.dropna(how='any')# to drop row having atleast on Nan value
```

```python
csv.dropna(how='all')# to drop row having all  Nan value
```

```python

import pandas as pd
csv=pd.read_csv('IRIS.csv')    #reading csv file

csv.dropna(subset=['species'])# to drop nan value from a perticular column using subset 
```

```python

import pandas as pd
csv = pd.read_csv('IRIS.csv')  # reading csv file

# to drop nan valuefrom dataset and replacing with some value using inplace
csv.dropna(inplace=True)
csv
```

```python

import pandas as pd
csv=pd.read_csv('IRIS.csv')    #reading csv file

csv.dropna(thresh=2)# to drop nan value from a perticular column having no of Nan value =2 or it can be any no
```

                            fillna

```python

import pandas as pd
csv=pd.read_csv('IRIS.csv')    #reading csv file

csv
```

```python

import pandas as pd
csv=pd.read_csv('IRIS.csv')    #reading csv file

csv.fillna('Python')# filling all NaN value to Python
```

```python

import pandas as pd
csv=pd.read_csv('IRIS.csv')    #reading csv file

csv.fillna({'species':'python','sepal_width':'25','sepal_length':50})# filling  NaN value of diffrent column with differnt value
```

```python
# filling  NaN value with forward  filling 
import pandas as pd
csv=pd.read_csv('IRIS.csv')    #reading csv file

csv.fillna(method='ffill',axis=1) # axiz=1 means it will fillede with column data  & 0 means row data
```

```python
# filling  NaN value with forward  filling 
import pandas as pd
csv=pd.read_csv('IRIS.csv')    #reading csv file

csv.fillna(method='bfill',axis=0)
```

```python
# filling  NaN value with forward  filling 
import pandas as pd
csv=pd.read_csv('IRIS.csv')    #reading csv file


csv.fillna(method='bfill',axis=1)
```

```python
# filling  NaN valueusing inplace
import pandas as pd
csv=pd.read_csv('IRIS.csv')    #reading csv file

csv.fillna('python',inplace=True)
csv
```

```python
# filling  NaN value with forward  filling 
import pandas as pd
csv=pd.read_csv('IRIS.csv')    #reading csv file

csv.fillna('python',limit=1)# it will limit upto 1 in each column
```
