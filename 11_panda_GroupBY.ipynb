                GroupBy

```python
import pandas as pd
import random as rd
s2=[rd.randint(0,50) for i in range (12)]
s3=[rd.randint(0,50) for i in range (12)]
var1=pd.DataFrame({'name':['a','b','c','d','c','a','b','a','c','b','a','c'],'S_1':[12,15,25,6,32,54,1,2,10,11,18,31],'S_2':s2,'S_3':s3})
var1
```

```python
var2=var1.groupby('name')
var2
```

```python
for i ,j in var2:
    print(i,j)
    print()
```

```python
var2.get_group('a')# to get only perticular data
# here only data of name a is appearing
```

```python
var2.min() # we get min data of each name like min of a ,b ,c
```

```python
var2.max() # to get maximum of each name
```

```python
var2.mean() # to get meam of each name
```
