```python
import pandas as pd
var=pd.DataFrame({'A':[1,2,3,4,5],'B':[1,2,3,4,5]})
var.insert(1,'new col',[5,4,3,2,1])  # inserting colum in dataframe
var
```

```python
import pandas as pd
var=pd.DataFrame({'A':[1,2,3,4,5],'B':[1,2,3,4,5]})
var.insert(1,'new col',var['A'])
var
```

```python
import pandas as pd
var=pd.DataFrame({'A':[1,2,3,4,5],'B':[1,2,3,4,5]})
var.insert(1,'new col',var['A'][:4]*2)

var
```
